```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv(r'C:\Users\sumanth\Downloads\datasetb2d9982\dataset\train.csv')
```

```python
train['PRODUCT_ID'].value_counts()
```

```python
tr=train[0:10000]
tr
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
tr.fillna(tr.TITLE)
tr.fillna(tr.BULLET_POINTS)
tr.fillna(tr.DESCRIPTION)
tr.fillna(tr.PRODUCT_TYPE_ID)
tr.head()
```

```python
card_docs_title_train = [TaggedDocument(str(doc).split(' '), [i]) for i, doc in enumerate(tr.TITLE)]
model = Doc2Vec(vector_size=64, window=2, min_count=1, workers=8, epochs = 40)
model.build_vocab(card_docs_title_train)
model.train(card_docs_title_train, total_examples=model.corpus_count, epochs=model.epochs)
card_docs_title_train
```

```python
card2vec_title_train = [model.infer_vector((tr['TITLE'][i].split(' '))) 
            for i in range(0,len(tr['TITLE']))]
card2vec_title_train 
```

```python
dtv= np.array(card2vec_title_train).tolist()
#set list to dataframe column
tr['TITLE'] = (dtv)
tr
```

```python
for i in tr.index:
    tr.TITLE[i]=sum(tr.TITLE[i])
```

```python
card_docs_bullet_train = [TaggedDocument(str(doc).split(' '), [i]) for i, doc in enumerate(tr.BULLET_POINTS)]
model = Doc2Vec(vector_size=64, window=2, min_count=1, workers=8, epochs = 40)
model.build_vocab(card_docs_title_train)
model.train(card_docs_bullet_train, total_examples=model.corpus_count, epochs=model.epochs)
card2vec_bullet_train = [model.infer_vector((str(tr['BULLET_POINTS'][i]).split(' '))) 
            for i in range(0,len(tr['BULLET_POINTS']))]
dtv= np.array(card2vec_bullet_train).tolist()
#set list to dataframe column
tr['BULLET_POINTS'] = dtv
card_docs_des_train = [TaggedDocument(str(doc).split(' '), [i]) for i, doc in enumerate(tr.DESCRIPTION)]
model = Doc2Vec(vector_size=64, window=2, min_count=1, workers=8, epochs = 40)
model.build_vocab(card_docs_title_train)
model.train(card_docs_des_train, total_examples=model.corpus_count, epochs=model.epochs)
card2vec_des_train = [model.infer_vector((str(tr['DESCRIPTION'][i]).split(' '))) 
            for i in range(0,len(tr['DESCRIPTION']))]
dtv= np.array(card2vec_des_train).tolist()
#set list to dataframe column
tr['DESCRIPTION'] = dtv
tr
```

```python
for i in tr.index:
    tr.DESCRIPTION[i]=sum(tr.DESCRIPTION[i])
    tr.BULLET_POINTS[i]=sum(tr.BULLET_POINTS[i])
```

```python
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
```

```python
Xtrain=tr.drop(['PRODUCT_LENGTH','PRODUCT_ID','PRODUCT_TYPE_ID'],axis=1)
Ytrain=tr['PRODUCT_LENGTH']
xTrain,xTest,yTrain,yTest=train_test_split(Xtrain,Ytrain,test_size=0.01,random_state=101)
```

```python
regressor=RandomForestRegressor(n_estimators=100,random_state=101)
```

```python
regressor.fit(xTrain,yTrain)
```

```python
y_pred=regressor.predict(xTest)
```

```python
#y_pred=pd.DataFrame(y_pred,columns=['yPRED'])
y_pred
```

```python
regressor.score(xTrain,yTrain)
```

```python
metrics.mean_absolute_percentage_error(yTest,y_pred)
```
